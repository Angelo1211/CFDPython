## Effect of Grid size on convergence


```python
# Adding inline command to make plots appear under comments
import numpy as np
import matplotlib.pyplot as plt
import time, sys
%matplotlib inline   

#General simulation initial conditions
grid_length = 2
nt = 20
dt = 0.025  #delta t
c  = 1 
```

```python
def linearConv(grid_points):
    dx = grid_length / (grid_points - 1) 
    
    #Initiallizing the shape of the wave to the same one from step 1 and displaying it
    u = np.ones(grid_points)
    u[int(.5/ dx):int(1 / dx + 1)] = 2
    
    un = np.ones(grid_points)
    
    for n in range(nt): #Runs however many timesteps you set earlier
        un = u.copy()   #copy the u array to not overwrite values
        for i in range(1,grid_points):
            u[i] = un[i] - c * dt/dx * (un[i]-un[i-1]) 
            
    plt.plot(np.linspace(0,grid_length,grid_points), u);
```

## What happened?

In the beginning, the increase in grid points was being pretty beneficial and was reducing some of the numnerical diffusion we were seeing. This made the wave approach more the square we had originally set to produce. 

However, after a certain point we decreased `dx` to the point that it was smaller than the distance covered by the wave during $ \Delta t $. This caused an unstable solution to occur and resulted in the nonsense that we have above. To correct for this we can enforce a stable solution by calculating a $ \Delta t$  that is calculated with respect to `dx` and in that way ensure that it is not too small in relation to thes step size. 

### The Courant Number

$$ \sigma = \frac{u \Delta t}{\Delta x} \leq \sigma_{max} $$

$\sigma$ is what is know as the **Courant Number** and the value of $\sigma_{max}$ is what represents the upper threshold of stability. This value depends on the discretization method used.

We shall now modify the function for linear convection to use the appropriate `dt` based on the size of `dx`.

```python
def linearConv2(grid_points):
    #Initial conditions
    sigma = 0.5
    dx = grid_length / (grid_points - 1) 
    c = 1 
    
    #Correcting dt for numerical stability reasons.
    dt = sigma * dx / c
    
    #Initiallizing the shape of the wave to the same one from step 1 and displaying it
    u = np.ones(grid_points)
    u[int(.5/ dx):int(1 / dx + 1)] = 2
    
     
    
    un = np.ones(grid_points)
    
    for n in range(nt): #Runs however many timesteps you set earlier
        un = u.copy()   #copy the u array to not overwrite values
        for i in range(1,grid_points):
            u[i] = un[i] - c * dt/dx * (un[i]-un[i-1]) 
            
    plt.plot(np.linspace(0,grid_length,grid_points), u);
```
